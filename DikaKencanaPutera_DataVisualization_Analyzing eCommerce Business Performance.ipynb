```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
```

```python
df = pd.read_csv('E:/DIKA/DATA PRODUKTIF DIKA/DATA SCIENCE/Rakamin/Mini project/Analyzing eCommerce Business Performance with SQL/Tugas 2/query_result.csv')
```

```python
df
```

```python
cat = []
num = []
for col in df.columns: 
    if df[col].dtype == 'O':
        cat.append(col)
    else: 
        num.append(col)
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
plt.plot(df.year, df.total_new_customer, color ='blue', linewidth=2.5, marker='o')
plt.title('New Customers Over Year', fontsize = 14, pad=20, weight='bold')
plt.xlabel('Year')
plt.ylabel('Total Customer')
plt.xticks(range(2016, 2019))
sns.despine()
plt.grid(color='darkgray', linestyle=':')
plt.tight_layout()
plt.savefig('NewCustomers.png')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
plt.plot(df.year, df.total_repeat_customer, color ='red', linewidth=2.5, marker='o')
plt.title('Repeat Order Customers Over Year', fontsize = 14, pad=20, weight='bold')
plt.xlabel('Year')
plt.ylabel('Total Customer')
plt.xticks(range(2016, 2019))
sns.despine()
plt.grid(color='darkgray', linestyle=':')
plt.tight_layout()
plt.savefig('RepeatOrderCustomers.png')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
plt.plot(df.year, df.avg_mau, color ='green', linewidth=2.5, marker='o')
plt.title('Average MAU (Monthly Active Users)\nOver Year', fontsize = 14, pad=20, weight='bold')
plt.xlabel('Year')
plt.ylabel('Average Customer')
plt.xticks(range(2016, 2019))
sns.despine()
plt.grid(color='darkgray', linestyle=':')
plt.tight_layout()
plt.savefig('MAU.png')
plt.show()
```
